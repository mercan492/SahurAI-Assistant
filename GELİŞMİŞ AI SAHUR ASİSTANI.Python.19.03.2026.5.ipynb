```python
import random

def sahur_asistani():
    # Veri tabanı: Besinler, Kaloriler (100g/adet) ve Protein miktarları (gram)
    besin_verileri = {
        "yumurta": {"kalori": 77, "protein": 6, "not": "Yüksek biyoyararlanım."},
        "lor peyniri": {"kalori": 100, "protein": 18, "not": "Yavaş sindirilen kazein."},
        "yulaf": {"kalori": 350, "protein": 12, "not": "Kompleks karbonhidrat."},
        "yogurt": {"kalori": 60, "protein": 3, "not": "Probiyotik destekli protein."},
        "hindi fume": {"kalori": 110, "protein": 20, "not": "Yağsız protein kaynağı."},
        "ceviz": {"kalori": 650, "protein": 15, "not": "Sağlıklı yağlar ve omega-3."}
    }

    print("---  GELİŞMİŞ AI SAHUR ASİSTANI v2.0 ---")

    # 1. Kullanıcı Girişi (Buzdolabı Taraması)
    print("\nDolabında neler var? (Lütfen aralarına virgül koyarak yazın)")
    print(f"Mevcut seçenekler: {', '.join(besin_verileri.keys())}")
    input_besinler = input("Giriş: ").lower().replace(" ", "").split(",")

    # Mevcut olanları filtrele
    mevcutlar = [b for b in input_besinler if b in besin_verileri]

    if not mevcutlar:
        print(" Üzgünüm, veri tabanımda eşleşen bir malzeme bulamadım. Örnek: 'yumurta, ceviz' yazabilirsin.")
        return

    # 2. Besin Analizi ve Hesaplama
    toplam_protein = 0
    toplam_kalori = 0
    
    print("\n SEÇİLEN MENÜ ANALİZİ:")
    for b in mevcutlar:
        p = besin_verileri[b]["protein"]
        k = besin_verileri[b]["kalori"]
        notu = besin_verileri[b]["not"]
        toplam_protein += p
        toplam_kalori += k
        print(f"- {b.capitalize()}: {p}g Protein | {k} kcal | {notu}")

    print("-" * 30)
    print(f" Toplam Tahmini Protein: {toplam_protein}g")
    print(f" Toplam Tahmini Kalori: {toplam_kalori} kcal")

    # 3. Su Hatırlatıcı Fonksiyonu
    try:
        kilo = float(input("\n Kilonuzu girin (Su hesabı için): "))
        # Bilimsel su ihtiyacı: Kilo * 0.035 (Litre)
        su_ihtiyaci = kilo * 0.035
        print(f" Sahur ve iftar arası toplam içmen gereken: {su_ihtiyaci:.2f} Litre.")
        print(f" Sahurda şu an en az {su_ihtiyaci/4:.2f} Litre içmelisin.")
    except ValueError:
        print(" Geçerli bir kilo girilmediği için su hesabı atlandı.")

    print("\n AI Önerisi: Bu menü seni yaklaşık 14 saat tok tutacaktır. Afiyet olsun!")

if __name__ == "__main__":
    sahur_asistani()
    
```
